```python
import pandas as pd

uscities = pd.read_csv('uscities.csv')
df = pd.read_csv('../../data_collection/plein_de_data/fandom_traitées.csv')
t=0
for state in df.State.unique():
    print(state)
    s=0
    for city in df[df.State == state].Town.unique():
        if city not in uscities[uscities.state_id == state].city.unique():
            print(city)
            s+=1
    print(s)
    t+=s
print(t)
```

```python
uscities = pd.read_csv('uscities.csv')
df = pd.read_csv('../../data_collection/plein_de_data/fandom_traitées.csv')

pbic_cities = {}

for state in uscities.state_id.unique():
    s = uscities[uscities.state_id == state].groupby('city').count()
    pbic_cities[state] = list(s[s.city_ascii >=2].index.unique())

for state in uscities.state_id.unique():
    for town in pbic_cities[state]:
        if town in df[df.State == state].Town.unique():
            print(state)
            print(town)
            print('================================================================')
```
