```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df15 = pd.read_csv('../../Nielsen/aggregated_nielsen_2015.csv')
df16 = pd.read_csv('../../Nielsen/aggregated_nielsen_2016.csv')
df15['year'] = 2015
df16['year'] = 2016

df = pd.concat((df15, df16))
```

```python
treat = 12029
control = 12079
```

```python
categories = [
    "FRESH PRODUCE",
    "BREAD AND BAKED GOODS",
    "MILK",
    "SNACKS",
    "PACKAGED MEATS-DELI",
    "CHEESE",
    "CARBONATED BEVERAGES",
    "CONDIMENTS, GRAVIES, AND SAUCES"
]

nwm = df[~df.is_walmart].copy()
nwm['purchase0'] = nwm.purchase_month + 12 *(nwm.purchase_year == 2016)



for category in categories:
    try:
        plt.figure(figsize=(10, 5))
        print(category)
        plt.plot(nwm[(nwm.guessed_store_county_fips==treat)&(nwm.product_group_descr==category)].sort_values('purchase0').purchase0, nwm[(nwm.guessed_store_county_fips==treat)&(nwm.product_group_descr==category)].sort_values('purchase0').upc_price, color='red')
        plt.plot(nwm[(nwm.guessed_store_county_fips==control)&(nwm.product_group_descr==category)].sort_values('purchase0').purchase0, nwm[(nwm.guessed_store_county_fips==control)&(nwm.product_group_descr==category)].sort_values('purchase0').upc_price, color='black')
        plt.show()
    except:
        pass
```
